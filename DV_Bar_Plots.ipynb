# **Bar Plots**

```python
#Librerias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#barplot / barras
sns.barplot(data=titanic, x="sex", y="age", estimator='mean', errorbar=None, palette='viridis')
plt.title('Barplot')
plt.xlabel('Sexo')
plt.ylabel('Promedio de edad')
plt.show()
```

```python
#Barras appiladas
data1 = titanic.groupby('class').agg({'alive':'count', 'sex': lambda x:(x=='male').sum()}).reset_index()
data1.rename(columns={'alive':'total', 'sex':'male'}, inplace=True)
data1.sort_values('total', inplace=True)
data1
```

```python
sns.set_color_codes("pastel")
sns.barplot(x="total", y="class", data=data1, label="Female", color="b")
sns.set_color_codes("muted")
sns.barplot(x="male", y="class", data=data1, label="Male", color="b")
plt.title("Barras apiladas")
plt.xlabel("Numero de pasajeros")
plt.legend(loc="upper right")
plt.show()
```

```python
#Barras agrupadas
sns.barplot(data=titanic, x='class', y='age', hue='sex', estimator='mean', errorbar=None, palette='magma')
plt.title('Barras agrupadas')
plt.xlabel('Clase')
plt.ylabel('Promedio de edad')
plt.show()
```
